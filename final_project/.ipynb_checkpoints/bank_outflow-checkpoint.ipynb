Создадим функцию, которая будет выводить информацию по датафрейму

```python
# создали функцию для просмотра и вывода информации датафрейма
def info(df):
    
    """ Функция выводит первые 10 строк датафрейма,
    выводит информацию по датафрейму в разрезе: количество строк, названия столбцов, 
    типы данных, есть ли пропуски в ячейках, проверяет на явные дубликаты,
    строит распределения признаков датафрейма"""
    
    display(df.head(10))
    print(df.info(), '*'*50, sep='\n')
    print(df.duplicated().sum(), '*'*50, sep='\n')
    plt.figure(figsize=(12, 12))
    plt.suptitle('Рапределение признаков датафрейма', fontsize=10)
    n=1
    for col in bank.drop('USERID', axis=1).columns:
        plt.subplot(4,4, n)
        plt.rc('font', size=8)
        sns.histplot(bank[col])
        sns.despine()
        plt.ylabel('')
        plt.xlabel(col)
        n+=1
    plt.show();
```

```python
# применили функцию к переменной bank
info(bank)
```

В датафрейме 10 тыс.строк и 12 столбцов. Названия некоторых столбцов записаны большими буквами, исправим для простоты использования. - баллы кредитного скоринга распределны от 600 до 1000, явных выбросов нет
- возраст рапределен от 18 до 90, явных выбросов нет,

Посмотрели на пропуски в столбце balance

```python
#вывели пропуски в столбце balance
bank[bank['balance'].isna()]
```

Есть гипотеза, что пропуски в столбцах balance связаны, что у клиента есть только кредитная карта и больше никаких других продуктов. Баланс кредитной карты не идет в общий зачет средств на счетах у клиента. Проверим данную гипотезу. Закодируем  значения как 1 - есть данные, пропуски как - 0. Для этого создадим новый столбец balance_kod.

```python
#создали новый столбец и закодировали пропуски
bank['balance_kod'] = bank['balance'].isna().map({True: 0, False: 1})
bank['balance_kod'].head()
```

```python
# сгруппировали по столбцу balance_kod - количество клиентов итого, количество продуктов на клиенте, сколько кредитных карт
balance_kod = (
    bank
    .groupby('balance_kod', as_index=False)
    .agg({'age': 'count','products': 'median', 'credit_card': 'sum'})
)
balance_kod
```

```python
print(
    'Доля клиентов без баласов с кредитными картами - {:.2%}'
    .format(balance_kod['credit_card'][0] / balance_kod['age'][0])
)
print(
    'Количество клиентов без баланса и кредитной карты - {}'
    .format(balance_kod['age'][0] - balance_kod['credit_card'][0])
)
```

```python
# посмотрели на клиентов без кредитной карты и баланса
bank[(bank['balance_kod'] == 0) & (bank['credit_card'] != 1)]
```

Гипотеза практически подтвердилась - 81,5% клиентов без баланса имеют только кредитную карту. Остальные 421 клиент имеют какой-то другой продукт банка. В данном датафрейме нет разбивки по остальным продуктам, только их количество у клиента. Можно предположить, что это может быть либо кредит, либо продукты рассрочки.

```python
# посмотрели на средний кредитный бал по данным клиентам
'Средний балл кредитного скоринга - {:.1f}'.format(
    bank[(bank['balance_kod'] == 0) & (bank['credit_card'] != 1)]['score'].mean()
)
```

```python
# посмотрели на отток данных клиентов.
'Отток клиентов без баласа и кредитной карты - {:.2%}'.format(
    bank[(bank['balance_kod'] == 0) & (bank['credit_card'] != 1)]['churn'].mean()
)
```

У клиентов без кредитной карты и без баланса высокий кредитный балл, говорит о том, что часто берут и погашают кредиты. Отток в размере 0.48%, в данном исследовании интересуют клиенты с высоким оттоком. Можем выделить что кредитная карта и кредит являются якорными продуктами и снижают отток.

```python
#сформировали признаки по банку по средней
bank_mean = (bank
 .groupby('portrait')
 .mean()
).T.reset_index()
bank_mean.columns = ['', 'bank_mean']

#сформировали признаки по банку по медиане
bank_median = (bank
 .groupby('portrait')
 .median()
).T.reset_index()
bank_median.columns = ['', 'bank_median']

#сформировали признаки по отточным клиентам по средней
outflow_mean = (outflow
 .groupby('portrait')
 .mean()
).T.reset_index()
outflow_mean.columns = ['', 'outflow_mean']

#сформировали признаки по отточным клиентам по медиане
outflow_median = (outflow
 .groupby('portrait')
 .median()
).T.reset_index()
outflow_median.columns = ['', 'outflow_median']

#сформировали признаки по не отточным клиентам по средней
non_outflow_mean = (non_outflow
 .groupby('portrait')
 .mean()
).T.reset_index()
non_outflow_mean.columns = ['', 'non_outflow_mean']

#сформировали признаки по не отточным клиентам по медиане
non_outflow_median = (non_outflow
 .groupby('portrait')
 .median()
).T.reset_index()
non_outflow_median.columns = ['', 'non_outflow_median']

#объединили все группировки в одну таблицу
portrets = np.round(
    reduce(lambda left,right: pd.merge(left,right,on=[''],
    how='outer'), 
    [bank_mean, 
     bank_median, 
     outflow_mean, 
     outflow_median, 
     non_outflow_mean, 
     non_outflow_median]
          ), 2
)

portrets
```

## Проверка статистических гипотез

В процессе анализа мы нашли признаки, которые коррелируют с целевой переменной, нашли в них интервалы, где отток преобладает. Чтобы двигаться далее для определения сегментов по оттоку и составления рекомендаций, проверим что отличия у признаков у отточных и не отточных клиентов являются статистически достоверными.

У нас есть датафрейм не отточных клиентов.

```python
#вывели 5 строк не отточных клиентов
non_outflow.head()
```

Есть датафрейм отточных клиентов.

```python
#вывели 5 строк отточных клиентов
outflow.head()
```

Зададим уровень статистической значимости на уровне 5%.

```python
alpha = 0.05
```

Проверим гипотезы различий в доходе, балансе и доле кредитных карт между теми клиентами, которые ушли и теми, которые остались

<a id='3.1-bullet'></a>
### Проверка гипотезы различия дохода между теми клиентами, которые ушли и теми, которые остались

Распределение заработной платы далеко от нормального распредления - есть выбросы, игнорировать которые нельзя. Алгебраические метрики в данном случае могут плохо работать. Используем тест, основанный на структурном подходе, или непараметрический. Метод,
который будем применять  U-критерий Манна-Уитни.

Сформируем нулевую и альтернативную гипотезу:  
HO - Доход не отличается между теми клиентами, что ушли и теми, которые остались  
Н1 - Доход клиетов, которые ушли, отличается у тех, которые остались

```python
# применили U-критерий Манна-Уитни
results = st.mannwhitneyu(non_outflow['est_salary'], outflow['est_salary'])
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
```

```python
print('Cредний доход у не отточных клиентов - {:.0f} рублей'.format(non_outflow['est_salary'].mean()))
print('Cредний доход у отточных клиентов    - {:.0f} рублей'.format(outflow['est_salary'].mean()))
```

U-критерий отверг нулевую гипотезу. Если смотреть на средние у отточных и не отточных клиентов, разница в 682 рубля.
Если еще раз посомтреть на распредление показателя, то у отточных и не отточных клиентов оно совпадает.

Промерим гипотезу с помощью еще  одного метода - двухвыборочный t-теста для независимых выборок 

```python
#применили ttest
results = st.ttest_ind(non_outflow['est_salary'], outflow['est_salary'])
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
```

Данный тест уже показал совсем другой результат - нельзя сделать вывод о различии. В матрице корреляции доход также коррелировал  на низком уровне - 5%. Однозначного вывода по данному показателю сделать нельзя, нужно смотреть во взаимодействии с другими признаками на уровень оттока.

<a id='3.2-bullet'></a>
### Проверка гипотезы различия баланса между теми клиентами, которые ушли и теми, которые остались

Ранее обнаружили, что балас очень сильно коррелирует с оттоком и с хорошими балансами, склонны к уходу чаще, чем остальные.

Сформируем нулевую и альтернативную гипотезу:  
HO - Баланс не отличается между теми клиентами, что ушли и теми, которые остались  
Н1 - Баланс клиетов, которые ушли отличается у тех, которые остались

Распределени у данного показателя не нормальное, однако совпадает для отточных и не отточных клиентов. Проверим данную гипотезу как и прошлую с помощью 2 тестов.


```python
# применили U-критерий Манна-Уитни
results = st.mannwhitneyu(non_outflow.query('balance_kod != 0')['balance'], outflow.query('balance_kod != 0')['balance'])
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
```

```python
#применили ttest
results = st.ttest_ind(non_outflow.query('balance_kod != 0')['balance'], outflow.query('balance_kod != 0')['balance'], alternative='less')
print('p-значение: ', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу: разница статистически значима")
else:
    print("Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя")
```

```python
print('Баланс у не отточных клиентов - {:.0f}'.format(non_outflow.query('balance_kod != 0')['balance'].mean()))
print('Баланс у отточных клиентов    - {:.0f}'.format(outflow.query('balance_kod != 0')['balance'].mean()))
print('Баланс отточных клиентов превышает не отточных в {:.1f} раза'
      .format(outflow.query('balance_kod != 0')['balance'].mean() / 
              non_outflow.query('balance_kod != 0')['balance'].mean()))
```

Нулевую гипотезу отверги оба теста и по средним мы видим разницу превышения в 1.5 раза, баласа у клиентов которые ушли.

<a id='3.3-bullet'></a>
### Проверка гипотезы влияния кредитной карты на клиентов, которые ушли и на тех, которые остались

Ранее обнаружили, что кредитная карта у клиентов является якорным продуктом, у тех у кого она есть уходят реже. Проверим данную гипотезу.

Сформулируем нулевую и альтернативные гипотезы:   
HO - Доля кредитных карт не отличается между теми клиентами, что ушли и теми, которые остались  
Н1 - Доля кредитных карт, у клиентов которые ушли отличается от тех, которые остались

Гипотезу будет проверять с помощью ztesta - метод проверки гипотез для пропорции, или доли. Стандартного теста в базовых
библиотеках Python нет, напишем его сами.

```python
# напишем функцию, которая будет считать доли пользователей
# и проверять будет ли отличие между группами статистически достоверным

def ztest(sign, non_outflow, outflow, alpha):
    alpha = alpha
    
    a1 = non_outflow[sign].sum()
    a2 = outflow[sign].sum()
    b1 = non_outflow[sign].count()
    b2 = outflow[sign].count()
    
    p1 = a1 / b1
    p2 = a2 / b2
    
    p_combined = (a1 + a2) / (b1 + b2)
    
    difference = p1 - p2
   
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/b1 + 1/b2))
    
    distr = st.norm(0, 1)
    
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    
    print('Признак -', sign)
    print('Количество пользователей, совершивших событие:')
    print('Группа: не отток - ', a1, 'пользователей с признаком', '/', 
          np.round(p1*100,2), '% - доля пользователей от общего количества в группе')
    print('Группа: отток - ', a2, 'пользователей с признаком', '/', 
          np.round(p2*100,2), '% - доля пользователей от общего количества в группе')
    print()
    print('Нулевая гипотеза - доли одинаковые')
    print('Альтернативная гипотеза - доли разные')
    print()
    print('Уровень значимости alpha: ', alpha)
    print('p-значение: ', p_value)
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    )
```

```python
ztest('credit_card', non_outflow, outflow, alpha)
```

Статистически доказали, что отток у клиентов с кредитными картами ниже. Чем выше доля таких клиентов, тем ниже отток.

Данную функцию мы можем применить для проверки активности клиентов и статистически проверить, действительно ли активные клиенты уходят чаще, чем те, которые остались.

<a id='3.4-bullet'></a>
### Проверка гипотезы влияние активности клиентов, на тех которые ушли и тех, которые остались

Ранее также заметили, что активные клиенты уходят чаще, проверим так ли это.

Сформулируем нулевую и альтернативные гипотезы:   
HO - Доля активных клиентов не отличается между теми клиентами, что ушли и теми, которые остались  
Н1 - Доля активных клиентов, которые ушли отличается от тех, которые остались

```python
ztest('last_activity', non_outflow, outflow, alpha)
```

Альтернативная гипотеза подтвердилась доли активных клиентов разные в датафреймах. 70% клиентов в оттоке были активными до ухода из банка.

На данном этапе проверили 4 гипотезы. Установили, что у признаки: баланс, активность клиентов и наличие кредитной карты - статистически достоверно влияют на отток клиентов. Признак доход клиента остается под вопросом, разные тесты показывают противополный результат.

## Вывод

Для проведения исследования был получен датасет  с данными о клиентах банка. Необходимо было проанализировать клиентов регионального банка и выделить сегменты клиентов, которые склонны уходить из банка.

В датасете проведена работа по замене названий столбцов - привели к нижнему регистру, обработаны дубликаты - нашли одинаковые userid клиентов, пропуски и типы данных - удалили пропуски в столбце возраст, закодировали пропуски в балансе, изменили тип данных в столбце возраст и скоринговый бал клиента.  

Провели исследовательский анализ данных, определили все значимые признаки отточности: интервалы значений характеристик, которые связаны с повышенным оттоком, сравнили портреты типичных клиентов, которые склонны и не склонны уходить из банка. Было выявлено, что чем больше продуктов у клиента, чем выше его баланс и количество баллов собственности - тем выше его склонность к уходу. Также чаще уходят из банка активные клиенты, клиенты мужского пола, клиенты с высоким скоринговым баллом.  

Сформулировали и проверили статистические гипотезы для подтверждения достоверности отличий признаков у клиентов склонных к оттоку и тех, кто остается в банке. Так у клиентов с кредитной картой статистически достоверно склонность к уходу ниже. Также статистически достоверно, что клиенты с высокими балансами и активные уходят чаще. В отношении дохода сложилась неоднозная картина, разные тесты показывают разный результат.

Объединили признаки отточности, сформировали сегменты, отобрали из них лучшие. Среди сегментов оказались активные клиенты с количеством продуктов больше 2, премиальные клиенты с доходом и балансом больше 1 млн.руб. и баллами собственности больше 2, мужчины без кредитной карты и высокими скоринговыми баллами. По данным сегментам дали рекомендации по работе и снижению оттока.
3-я выделенными сегментами смогли охватить 57% отточных клиентов.  

По итогам исследования подготовили презентацию.
